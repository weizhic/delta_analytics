```python
def expand_lookups(user):
    user['expand'] = user.recentLookupCount.apply(lambda x: pd.DataFrame(x))
    df = pd.DataFrame(columns=['createdAt','wordstring','_id','recentLookups'])
    for i in user.index:
        user0 = user.loc[i,'expand']
        user0['_id'] = user.loc[i,'_id']
        user0['recentLookups'] = user.loc[i,'recentLookups']
        df = df.append(user0)
    return df
```

```python
ind = [] + ['-{}'.format(i) for i in xrange(2,11)]
files = ['/Users/wcheung/Dropbox/delta-analytics/data-19Jan2019/userlookups-01192019{}.json'.format(j) for j in ind]
```

```python
df = pd.DataFrame(columns=['createdAt','wordstring','_id','recentLookups'])
for f in files:
    user = pd.read_json(f)
    df_ = expand_lookups(user)
    df = df.append(df_)
```
